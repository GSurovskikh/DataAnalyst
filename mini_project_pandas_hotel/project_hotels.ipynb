```python
countries = df \
    .query('is_canceled != "0"') \
    
```

```python
#Пользователи из каких стран совершили наибольшее число успешных бронирований?
countries.country.value_counts()
```

```python
city_hotel_nights = df \
    .query("hotel == 'City Hotel'") \
    .agg({'stays_total_nights':'mean'})
```

```python
#На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel? 
city_hotel_nights.round(2)
```

```python
resort_hotel_nights = df \
    .query("hotel == 'Resort Hotel'") \
    .agg({'stays_total_nights':'mean'})
```

```python
#На сколько ночей (stays_total_nights)  в среднем бронируют отели типа Resort Hotel?
resort_hotel_nights.round(2)
```

```python
diff_type_room = df \
    .query("reserved_room_type != assigned_room_type")
```

```python
# Тип заброн. комнаты отличается от фактической 
diff_type_room
```

```python
#Группировка бронирований по месяцам и годам
df.groupby(['arrival_date_year','arrival_date_month'],as_index=False) \
    .agg({"is_canceled" : 'count'}) \
    .sort_values("is_canceled",ascending=False)
```

```python
#В каких месяцах отменялись брони чаще по годам
df.query("hotel=='City Hotel'").query("is_canceled == 1") \
    .groupby('arrival_date_year').arrival_date_month.value_counts()
    
```

```python
# Кого больше в отелях
df[["babies","children","adults"]].mean()
```

```python
df['total_kids']=df.babies + df.children
```

```python
#В каком типе отелей детей больше среднее значение
df.groupby("hotel").total_kids.mean()
```

```python
def has_kids(kid):
    if kid > 0:
        return True
    else:
        return False
```

```python
#Расчет процента отмены брони,в зависимости от наличия детей
df.groupby('has_kids')['is_canceled'].value_counts(normalize=True)
```
