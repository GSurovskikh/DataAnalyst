# Задача №2: Сколько заказов в месяц в среднем не доставляется по разным причинам (вывести детализацию по причинам)

```python
#Недоставленные заказы, те заказы у которых нет статуса "delivered"
orders_not_delivered = orders.query("order_status != 'delivered'") \
    .groupby("order_status",as_index=False) \
    .agg({'order_id':'count'}) \
    .rename(columns={'order_id':'not_delivered'})
```

```python
#Кол-во недоставленных заказов с разбивкой по статусу заказа
orders_not_delivered
```

# Задача №3: По каждому товару определить, в какой день недели товар чаще всего покупается.

```python
#Для нахождения времени заказа необходимо
#объеденить датафреймы товаров и заказов
popular_item_day = items.merge(orders,how='left',on='order_id')
```

```python
popular_item_day.head()
```

```python
#Для определения дня недели каждого заказа
#Необходимо создать колонку,которая будет отображать дату совершения заказа в днях недели(weekday)
popular_item_day['weekday'] = popular_item_day.order_purchase_timestamp.dt.day_name()
popular_item_day.head()
```

После объединения двух датафреймов,
делаем фильтрацию по интересующим нас параметрам.

А именно:

    Статус заказа - 'delivered',
    
    Количество заказов за день - максимальное,
    
    Сортировка данный по количеству заказов впорядке убывания,
    
    Сброс повторов наименований

```python
item_by_day = popular_item_day.query('order_status == "delivered"') \
                        .groupby(['product_id','weekday'],as_index=False) \
                        .agg({'order_id':'count'}) \
                        .sort_values(['order_id'],ascending=False) \
                        .drop_duplicates(subset=['product_id'],keep='first') \
                        .drop_duplicates(subset=['weekday'],keep='first') \
                        .rename(columns={'order_id':'purchase_count'})


item_by_day
```

Также из-за фильтрации на предыдущем шаге,
мы можем упустить тот момент, когда один и тот же товар является популярным в нескольких днях.
Для этого сформируем похожий запрос, но повторы сбрасываем только для дней недели

```python
item_by_day2 = popular_item_day.query('order_status == "delivered"') \
                        .groupby(['product_id','weekday'],as_index=False) \
                        .agg({'order_id':'count'}) \
                        .sort_values(['order_id'],ascending=False) \
                        .drop_duplicates(subset=['weekday'],keep='first') \
                        .rename(columns={'order_id':'purchase_count'})


item_by_day2
```

# Задание №5 : Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.

Метрика Retention(удержание) - отношение кол-ва клиентов(в конце временного периода) к кол-ву клиентов(в начале временного периода)
В нашем случае временной период задан по условию - 3месяца.
(В данном задании мы будем использовать данные ТОЛЬКО за 2017год, т.к. в исходных данных этот временной отрезок представлен целиком,а также придерживаемся ранее обговоренного понятия "покупка" - заказ, с отмеченной датой доставки)

```python
#Оставляем интересующие заказы
cohorts_df = total_df.query("order_delivered_customer_date >= '2017-01-01' and order_delivered_customer_date <= '2017-12-01'")[['customer_unique_id', 'order_delivered_customer_date']].loc[(total_df['order_status'] == 'delivered')].reset_index(drop=True)
```

```python
cohorts_df['order_period'] = cohorts_df['order_delivered_customer_date'].astype('datetime64[M]')
```

```python
cohorts_df =cohorts_df.set_index('customer_unique_id')
cohorts_df
```

```python
#Колонка с датой первой покупки каждого пользователя
cohorts_df['first_purchase'] = cohorts_df.groupby(level=0)['order_delivered_customer_date'].min().astype('datetime64[M]')
cohorts_df.reset_index(inplace=True)
cohorts_df
```

```python
cohorts_df.insert(len(cohorts_df.columns), 'total_orders', 0, allow_duplicates=False) 
```

```python
cohorts = cohorts_df.groupby(['first_purchase','order_period']) 
cohorts = cohorts.agg({'customer_unique_id': pd.Series.nunique,
                          'total_orders': pd.Series.count})
cohorts.rename(columns={'customer_unique_id': 'total_users'}, inplace=True)
cohorts
```

```python
def cohort_period(cohorts_df):
    cohorts_df['cohort_period'] = np.arange(len(cohorts_df))
    return cohorts_df
cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts
```

```python
cohorts['total_users'].unstack(level=-1, fill_value=None)
```

```python
  # переиндексируем DataFrame
cohorts.reset_index(inplace=True)
#поменяли местами cohort_period и first_purchase
cohorts.set_index(['first_purchase','cohort_period'], inplace=True)

cohort_group_size = cohorts['total_users'].groupby(level=0).first()
cohorts['total_users'].unstack(0)
user_retention = cohorts['total_users'].unstack(0).divide(cohort_group_size, axis=1)
user_retention.head()
```

```python
sns.set(style='ticks')
plt.figure(figsize=(22, 16))
plt.title('Cohorts: User Retention')
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.2%')
```

# Задание №6:Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. 

Для детального анализа RFM-сегмента разделим каждую метрику на 3 границы:

R - давность:

     1 - давние
   
     2 - "спящие" относительно недавние
     
     3 - недавние
     
F - частота покупок:

    1 - разовая
    
    2 - редкие
    
    3 - частые
    
М -сумма покупок:

    1 - низкий чек
    
    2 - средний чек
    
    3 - высокий чек

```python
#Из общего датафрейма отбираем интересующие нас заказы(доставленные)
delivered_purchases = total_df.dropna(subset=['order_approved_at']).query("order_status == 'delivered'")
delivered_purchases.head()
```

```python
#Определим последнюю дату в нашей базе данных
total_df.order_delivered_carrier_date.max() 
```

Для анализа метрики Requency(время от последней покупки пользователя до определенной даты) возьмем последнюю дату в нашей базе данных + 1 день(2018-09-12) 

```python
#Создадим точку сравнения (последняя дата обновления +1 день):
point = total_df.order_delivered_carrier_date.max() + dt.timedelta(days=1)
point
```

```python
purchases_by_customers = total_df.groupby([ 'customer_unique_id','order_id', 'order_delivered_carrier_date'])\
.agg({'price':"sum"})\
.reset_index()
purchases_by_customers.head()
```

```python
#Теперь необходимо рассчитать кол-во дней от даты покупки до даты наблюдения(2018-09-12):
purchases_by_customers['diff_days']=  purchases_by_customers['order_delivered_carrier_date'].apply(lambda time: (point - time).days)
purchases_by_customers.head()
```

```python
#Находим минимальную разницу в днях между точками покупки и наблюдения,суммируем кол-во заказов, а также находим общую сумму
#Также переименуем колонки 
df_rfm = purchases_by_customers.groupby(['customer_unique_id']) \
                                .agg({'diff_days':lambda days: days.min()\
                                ,'order_delivered_carrier_date': lambda x: len(x)
                                ,'price': sum})\
                                .reset_index() \
                                .rename(columns={'diff_days':'Recency','order_delivered_carrier_date':'Frequency','price':'Monetary'})
df_rfm.head()
```

```python
#Теперь необходимо разделим покупателей на сегменты,согласно полученным данным.
#Для этого определим квантили всех распределенний
#Мы будем разделять покупателей на 3 сегмента.
#(Чем больше данных сегментов,тем точнее полученная сегментация)
quantiles = df_rfm[['Recency','Frequency', 'Monetary']].quantile([.33,.66]).to_dict()
quantiles
```

```python
#Из-за специфики данных, большинство наших клиентов(>90%) совершили всего одну покупку,
#по этой причине сегментация по метрике Frequency будет для большинства едина. 
df_rfm.Frequency.value_counts()
```

```python
#Далее напишем функции для ранжирования всех покупателей, согласно полученных выше квантилей распределения
def recency_score(x):
    if x <= quantiles['Recency'][.33]:
        return 3
    elif x <= quantiles['Recency'][.66]:
        return 2
    else:
        return 1
    
    
def frequency_score(x):
    if x <= 1:
        return 1
    elif 1 < x <= 3:
        return 2
    else:
        return 3
    
def monetary_score(x):
    if x <= quantiles['Monetary'][.33]:
        return 1
    elif x <= quantiles['Monetary'][.66]:
        return 2
    else:
        return 3
    
```

```python
#Сведем данные в одну таблицу 
df_rfm['R'] = df_rfm['Recency'].apply(lambda x: recency_score(x))
df_rfm['F'] = df_rfm['Frequency'].apply(lambda x: frequency_score(x))
df_rfm['M'] = df_rfm['Monetary'].apply(lambda x: monetary_score(x))
df_rfm['RFM'] = df_rfm['R'].map(str) + df_rfm['F'].map(str) + df_rfm['M'].map(str)  
df_rfm.head()
```

```python
result_rfm = pd.DataFrame(df_rfm['RFM'].value_counts())
result_rfm
```
